Another way we can look at the contributions to OBIS is through the datasets. These are also described in metadata being served up by IPT servers in the network, harvested into OBIS as part of the integration process, and exposed through the API. This notebook just runs a basic dump of dataset titles, total records (for the group of taxa associated with a given EOV), and abstract. I'll fiddle with some better ways of packaging and working through this information to facilitate gap analysis.

```python
from IPython.display import display
import pegasus_functions as pf
```

```python
eov_packet = pf.get_eov_packet()
```

```python
for index,eov in enumerate(eov_packet):
    dataset_stats = pf.summary_stats_by_aphiaids(pf.get_worms_info(eov["name"])["valid_aphiaids"], summary_type="dataset")

    print(f'{eov["label"]} - Contributing Datasets')
    print("------------")
    for dataset in dataset_stats["results"]:
        print(dataset["title"], "Records:", dataset["records"])
        print(dataset["abstract"])
        print("************")
    print("==============================")
```
